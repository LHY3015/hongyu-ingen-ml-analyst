# Environment Check


```python
import sys
import platform

print("SYSTEM INFORMATION\n")
print(f"Platform: {platform.platform()}")
print(f"Python Version: {sys.version}")
```

```python
import numpy as np
import scipy
import pandas as pd
import sklearn
import torch
import matplotlib
import seaborn as sns
import shap
import pytest

import_results = {
    'numpy': {'version': np.__version__, 'path': np.__file__},
    'scipy': {'version': scipy.__version__, 'path': scipy.__file__},
    'pandas': {'version': pd.__version__, 'path': pd.__file__},
    'sklearn': {'version': sklearn.__version__, 'path': sklearn.__file__},
    'torch': {'version': torch.__version__, 'path': torch.__file__, 
              'cuda': f"CUDA available: {torch.cuda.is_available()}"},
    'matplotlib': {'version': matplotlib.__version__, 'path': matplotlib.__file__},
    'seaborn': {'version': sns.__version__, 'path': sns.__file__},
    'shap': {'version': shap.__version__, 'path': shap.__file__},
    'pytest': {'version': pytest.__version__, 'path': pytest.__file__}
}

print("IMPORT RESULTS\n")
print("-" * 50)
for lib_name, info in import_results.items():
    print(f"{lib_name:<15s} version: {info['version']}")
    if lib_name == 'torch':
        print(f"{'':15s} {info['cuda']}")
print("-" * 50)
```
